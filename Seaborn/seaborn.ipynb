# Distribution Plot:
- distplot -> this for one feature
- joinplot -> two feature
- pairplot -> more than two

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head(

)
```

Correlation with Heatmap

```python
df1=df.corr(method='spearman',numeric_only=True)
```

```python
df1
```

```python
sns.heatmap(df1)
```

# Join Plot

Univariate Analysis

```python
sns.jointplot(x="tip",y="total_bill",data=df)
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="reg")
```

# Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

# Dist plot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# Categorical Plot


- boxplot
- violinplot
- countplot
- bar plot

```python
sns.countplot("sex",data=df)
```

```python
sns.countplot(y="sex",data=df)
```

```python
sns.barplot(x="total_bill",y='sex',data=df)
```

```python
# box plot

sns.boxplot(x='smoker',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
#violin plot

sns.violinplot(x='smoker',y='total_bill',data=df,palette='rainbow')
```
