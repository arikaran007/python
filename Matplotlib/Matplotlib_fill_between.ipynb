```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/data(1).csv')
data.head()
```

```python
ages = data['Age']
dev_salary = data['All_Devs']
py_salary = data['Python']
js_salary = data['JavaScript']
```

```python
plt.plot(ages,dev_salary,color='#444444',linestyle='--',label='All_Devs')
plt.plot(ages,py_salary,label='Python')

plt.legend()


plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
plt.plot(ages,dev_salary,color='#444444',linestyle='--',label='All_Devs')
plt.plot(ages,py_salary,label='Python')

plt.fill_between(ages,py_salary,alpha=0.25)

plt.legend()


plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
plt.plot(ages,dev_salary,color='#444444',linestyle='--',label='All_Devs')
plt.plot(ages,py_salary,label='Python')

overall_median = 5732
plt.fill_between(ages,py_salary,overall_median,alpha=0.25)

plt.legend()


plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
plt.plot(ages,dev_salary,color='#444444',linestyle='--',label='All_Devs')
plt.plot(ages,py_salary,label='Python')

plt.fill_between(ages, py_salary, dev_salary,
                 where=(py_salary > dev_salary),
                 interpolate=True, alpha=0.25, label='Above Avg')

plt.fill_between(ages, py_salary, dev_salary,
                 where=(py_salary <= dev_salary),
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')

plt.legend()


plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
