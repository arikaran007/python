```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/data(4).csv')
df.head()
```

```python
age = df['Age']
dev_sal=df['All_Devs']
py_sal=df['Python']
js_sal=df['JavaScript']
```

```python
plt.plot(age,dev_sal)
plt.plot(age,py_sal,linestyle='dotted',color='#444444')
plt.plot(age,js_sal)

plt.title('salary of developers')
plt.xlabel('age')
plt.ylabel('salary')
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn')

fig1,ax1 = plt.subplots()
fig2,ax2=plt.subplots()
fig3,ax3=plt.subplots()

ax1.plot(age,dev_sal,label='All-devs')
ax2.plot(age,py_sal,linestyle='dotted',color='#444444',label='Python')
ax3.plot(age,js_sal,label='JavaScript')



ax1.set_title('salary of developers')

ax1.set_ylabel('salary')
ax1.legend()

ax2.set_ylabel('salary')
ax2.legend()

ax3.set_ylabel('salary')
ax3.set_xlabel('age')
ax3.legend()



plt.tight_layout()

plt.show()

```
